### **Implementing Normal Equation for Linear regression**
for simple linear datasets we can use normal equation to directly find the best values of weights and bias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=2*np.random.rand(100,1)
y=4+3*X+np.random.randn(100,1)
```

```python
sns.scatterplot(x=X.flatten(),y=y.flatten())
```

```python
X_new=np.insert(X,0,1,axis=1)
theta=np.linalg.inv(X_new.T.dot(X_new)).dot(X_new.T).dot(y)
```

```python
theta
```

```python
sns.scatterplot(x=X.flatten(),y=y.flatten())
plt.plot(X,X_new.dot(theta),color='red')
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X,y)
```

```python
lin_reg.intercept_
```
